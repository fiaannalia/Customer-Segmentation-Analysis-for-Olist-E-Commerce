## Background

resize olist (1).jpeg


Olist merupakan platform e-commerce yang didirikan pada tahun 2014 di Curitiba, Brazil. Olist adalah perusahaan yang menyediakan perangkat lunak berbasis cloud (SaaS) kepada pelaku bisnis (B2B), terutama yang bergerak di bidang retail, untuk membantu mereka menjual produk ke konsumen secara online [[1]](https://tracxn.com/d/companies/olist/__CMW0T0XDiabVcanXT6RE1RkX1lZRPOJNImBhisXLVxg#about-the-company). Olist menawarkan beberapa layanan pendukung bagi pelaku bisnis, seperti dukungan logistik untuk pengiriman yang andal, layanan finansial melalui pinjaman dan penggalangan dana, serta perangkat pemasaran untuk membantu para pelaku bisnis mempromosikan produk dan layanannya [[2]](https://canvasbusinessmodel.com/blogs/how-it-works/olist-how-it-works?srsltid=AfmBOoqGlQX9C9_LRUq39kfebFDNjFDGrjNKr51gupwV-2aCaeDO4g-i).

Cara kerja Olist cukup sederhana dan mudah digunakan. Penjual cukup mendaftarkan produk melalui dashboard Olist, lalu sistem otomatis menyalurkan produk ke berbagai marketplace. Seluruh proses mulai dari pengelolaan pesanan, sistem pembayaran, hingga dukungan pelanggan dikelola langsung melalui platform Olist.

Pendapatan Olist berasal dari beberapa sumber. Pertama, Olist mengenakan biaya langganan (*subscription fee*) kepada penjual berdasarkan paket layanan yang dipilih. Biaya langganan dimulai dari R$45.90 per bulan [[3]](https://olist.com/planos/). Kedua, Olist memperoleh pendapatan dari biaya transaksi berupa persentase tertentu dari setiap penjualan yang berhasil melalui platform sebesar 10% [[4]](https://medium.com/@fercanc/how-can-olist-increase-its-profit-ef58e3f274e). Selain itu, Olist juga menawarkan layanan tambahan seperti logistik, kredit usaha, dan iklan pemasaran yang menjadi sumber pendapatan tambahan. Olist juga mendapatkan pendapatan lainnya melalui iklan dan sponsor di platformnya, serta melalui monetisasi data yang berkaitan dengan perilaku pelanggan, tren pasar, dan kinerja penjualan [[2]](https://canvasbusinessmodel.com/blogs/how-it-works/olist-how-it-works?srsltid=AfmBOoqGlQX9C9_LRUq39kfebFDNjFDGrjNKr51gupwV-2aCaeDO4g-i).



## Problem Statement
Meskipun Olist telah menyediakan berbagai layanan digital yang mendukung pelaku usaha kecil dan menengah dalam menjual produknya secara online, jika layanan seperti promosi atau logistik diberikan secara merata tanpa mempertimbangkan kualitas dan kinerja penjual, hal ini dapat menyebabkan pemborosan sumber daya dan tidak berdampak signifikan terhadap peningkatan pendapatan.


Olist ingin meningkatkan efisiensi operasional dan pertumbuhan pendapatan dengan mengidentifikasi penjual-penjual yang berpotensi memberikan kontribusi tinggi terhadap penjualan dan kepuasan pelanggan. Dengan analisis yang tepat, Olist dapat mengetahui penjual yang aktif, responsif, dan memiliki performa baik, lalu memprioritaskan mereka untuk menggunakan layanan tambahan seperti promosi, logistik, dan fasilitas finansial. Dengan pendekatan ini, Olist dapat mengalokasikan sumber daya secara optimal, meningkatkan potensi penjualan, meningkatkan pelayanan pelanggan, dan meningkatkan revenue perusahaan.


## Stakeholders
1. **Company (Olist)**

    Sebagai penyedia platform, Olist bertanggung jawab menyediakan sistem yang efisien dan terintegrasi bagi penjual dan pembeli. Olist berkepentingan untuk meningkatkan performa layanan dan efisiensi operasional.

2. **Seller (Penjual/UKM)**
    
    Penjual memanfaatkan platform Olist untuk menjangkau lebih banyak pelanggan dan meningkatkan penjualan. Mereka berharap mendapatkan dukungan berupa promosi, pengelolaan proses pengiriman, serta akses ke layanan keuangan. Kinerja penjual secara langsung memengaruhi kepuasan pembeli dan kesuksesan platform Olist secara keseluruhan.
3. **Buyer (Konsumen)**

    Pembeli adalah pengguna akhir yang bertransaksi di penjual yang menggunakan platform Olist. Mereka menginginkan proses belanja yang praktis, pengiriman yang tepat waktu, serta produk yang sesuai dengan deskripsi. Tingkat kepuasan pembeli akan berdampak pada ulasan yang diberikan, pendapatan perusahaan, dan loyalitas terhadap platform.

## Goals

1. Meningkatkan efisiensi operasional dan peningkatan pendapatan perusahaan dengan mengidentifikasi penjual yang berpotensi.
2. Memberikan peluang yang lebih besar bagi penjual yang berpotensi untuk mendapatkan dukungan layanan tambahan seperti promosi, logistik, dan fasilitas finansial yang lebih lanjut
3. Mendorong pertumbuhan volume transaksi dan kualitas layanan dari para penjual untuk secara langsung berkontribusi pada peningkatan kinerja finansial dan keberhasilan bisnis perushaan

### 2-1. Transaction Fee Revenue

**Menghitung Growth Revenue per Bulan dari Tiap Cluster**

Pertumbuhan revenue per bulan dihitung menggunakan CAGR (Compound Annual Growth Rate). Metode ini mengukur rata-rata tingkat pertumbuhan revenue per bulan selama periode waktu tertentu, dihitung dari nilai awal hingga nilai akhir periode. Keunggulan CAGR adalah memberikan gambaran tren pertumbuhan jangka panjang yang stabil, tanpa terpengaruh fluktuasi data di tengah-tengah periode. Jadi, meskipun di beberapa bulan revenue naik turun, CAGR hanya fokus pada total pertumbuhan dari titik awal ke titik akhir, lalu dirata-rata secara proporsional per bulan.

```python
# Hitung CAGR (Compound Annual Growth Rate) per cluster
cagr_list = []

for cluster in df_avg_cluster_revenue['cluster'].unique():
    df_temp = df_avg_cluster_revenue[df_avg_cluster_revenue['cluster'] == cluster].reset_index(drop=True)
    start_value = df_temp.loc[0, 'avg_revenue']
    end_value = df_temp.loc[len(df_temp)-1, 'avg_revenue']
    n_periods = len(df_temp) - 1  # periode dalam bulan

    if start_value > 0 and n_periods > 0:
        cagr = ((end_value / start_value) ** (1/n_periods) - 1) * 100
    else:
        cagr = None

    cagr_list.append({'cluster': cluster, 'Growth (%)': round(cagr, 2)})

df_avgRevenue_Growth = pd.DataFrame(cagr_list)
df_avgRevenue_Growth
```

- Feedback: Minta alasan ketidakaktifan dari seller sebagai masukan untuk perbaikan layanan.

**Tujuan:**
- Re-activate & mencegah churn.
- Efisiensi Biaya: Mengaktifkan kembali penjual yang sudah ada jauh lebih hemat biaya dibandingkan dengan mengakuisisi penjual baru.
- Meningkatkan Pendapatan Olist: Setiap penjual yang berhasil diaktifkan kembali akan berkontribusi pada pendapatan Olist melalui komisi penjualan  dan biaya langganan bulanan.
- Meningkatkan Kesehatan Ekosistem: Basis penjual yang aktif akan menciptakan ekosistem marketplace yang lebih dinamis dan menarik bagi pembeli, yang pada gilirannya akan menarik lebih banyak penjual baru.

**Estimasi Campaign:**
- Frekuensi: 1x kirim email kepada seller yang pasif selama 12 bulan. Jika belum aktif juga, bisa dilakukan kembali per kuartal. [[1]](https://www.twilio.com/en-us/blog/insights/4-tips-for-sending-re-engagement-email-campaigns)
- Tingkat reaktivasi: Tingkat reaktivasi yang dianggap sangat baik dalam e-commerce adalah 10-15% dari pelanggan (penjual) yang tidak aktif. [[2]](https://www.alexanderjarvis.com/what-is-reactivation-rate-in-ecommerce/)
- Estimasi biaya:
    - Biaya utama untuk kampanye ini adalah perangkat lunak pemasaran email (ESP) yang bervariasi tergantung pada jumlah kontak.
    - Mengingat Klaster 2 memiliki 2.659 penjual, Olist dapat menggunakan paket yang mendukung jumlah kontak tersebut dengan estimasi biaya $18 atau 100 BRL [[3]](https://www.emailvendorselection.com/email-marketing-cost-pricing-guide/)
- Potensi kenaikan pendapatan: Reaktivasi pelanggan (penjual) yang ada dapat menghasilkan peningkatan pendapatan yang signifikan, dengan studi menunjukkan potensi kenaikan 25-95%. [[4]](https://www.alexanderjarvis.com/what-is-reactivation-rate-in-ecommerce/) 

**Potensi Pertumbuhan Pendapatan Olist:**

Berdasarkan hasil clustering analysis dataset Olist Brazilian e-commerce, strategi diskon upgrade paket langganan ditargetkan kepada dua segmen utama: **Potential Seller** dan **High-Value Seller**.

Strategi ini dirancang untuk mendorong seller di 2 cluster tersebut untuk meningkatkan paket langganan mereka ke tingkatan yang lebih tinggi. Dengan menawarkan diskon sementara, Olist bertujuan untuk memberikan insentif finansial yang menarik, memungkinkan penjual untuk merasakan manfaat fitur-fitur premium yang dapat meningkatkan volume penjualan, efisiensi operasional, dan profitabilitas. - **Periode:** 3 bulan untuk habit formation, memungkinkan seller beradaptasi dengan fitur baru dan merasakan dampak peningkatan transaksi. [[2]](https://onlinelibrary.wiley.com/doi/abs/10.1002/ejsp.674)

Strategi repricing cerdas dapat menghasilkan peningkatan margin hingga 4,3% [[8]](https://www.channelengine.com/en/blog/strategies-for-driving-profitable-growth-on-marketplaces)
